### Geometry mesh generation: SVW_ver1/test/test_Mesh_inter.py
This class, Mesh_inter, accept .msh file, and covers the file into an organized equal spaced mesh with the interpolated geometry. There are two main parts: 

 - Interpolation density N: For each triangular mesh, this method interpolates data points based on its three vertices.  
```python
Mesh_inter(gmsh_file).mesh_interpo(N)
```
<br> 
 - Reconstruction density M: Reduced the interpolated mesh into equal spaced unit solid angle with density M. 
```python
Mesh_inter(gmsh_file).L_grid(M)
```
